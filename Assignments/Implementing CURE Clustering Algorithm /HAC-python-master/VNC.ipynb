# Variability Neighbor Clustering

```python
%matplotlib inline
```

```python
from HACluster import VNClusterer
from sklearn.metrics import pairwise_distances
import numpy as np
```

```python
data = np.array([[2,4], [0,1], [1,1], [3,2], [4,0], [2,2], [8, 9], [8, 11]])
D = pairwise_distances(data)
```

```python
vnc = VNClusterer(D, 'ward')
vnc.cluster()
```

```python
labels = ['n'+str(i+1) for i in range(len(data))]
t = vnc.dendrogram.ete_tree(labels=labels)
```

```python
t.render("%%inline", h=10, units='in')
```
